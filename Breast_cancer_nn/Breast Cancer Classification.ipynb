```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#Loading the dataset from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame
```

```python
data_frame['class'] = breast_cancer_dataset.target
```

```python
data_frame
```

```python
#Number of rows and columns
data_frame.shape
```

```python
data_frame.info()
```

```python
#Check for any null values inside the data frame
data_frame.isnull().sum()
```

```python
#Statistical measures of the data frame
data_frame.describe()
```

```python
#Checking the distribution of the target variable
data_frame['class'].value_counts()
"""
1 means Benign
0 means Malignant
"""
```

```python
data_frame.groupby('class').mean()
```

```python
#Seperate the Features and the target variable
X = data_frame.drop(columns = 'class', axis = 1)
y = data_frame['class']
```

```python
print(X)
```

```python
print(y)
```

```python
#Splittng the data into training sets and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
print(X_train)
```

```python
print(X_train.shape, X_test.shape)
```

```python
#Standardize the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#Setting up the layers of the Neural Network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation = 'relu'),
    keras.layers.Dense(2, activation = 'sigmoid')
])
```

```python
#Compiling the Neural Network
model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy']  
)
```

```python
history = model.fit(X_train, y_train, validation_split = 0.1, epochs = 10)
```

```python
#Visualizing the accuracy

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'])
```

```python
#Visualizing the loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'])
```

```python
#Accuracy of the model on the test set

loss, accuracy = model.evaluate(X_test, y_test)
print(accuracy)
```

```python
y_pred = model.predict(X_test)
```

```python
#Converting the prediction probability to class labels
y_pred_labels = [np.argmax(i) for i in y_pred]
print(y_pred_labels)
```
